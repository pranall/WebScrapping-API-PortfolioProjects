2.c)

* Implemented a user-defined function that accepts any URL, retrieves the webpage content, validates the response status, and returns a fully initialized BeautifulSoup instance.

* Included two test cases for the function: one using a valid, accessible URL and another using a URL designed to return a 404 error to confirm proper exception handling.

```python
import requests
from bs4 import BeautifulSoup

def get_BS(url):
  needed_headers = {'User-Agent': "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36"}

  try:
    res = requests.get(url,headers=needed_headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.content,"html.parser")
    return soup

  except requests.exceptions.RequestException as e:
    print("Error : ", e)
    return None
```

4) User defined functions to return for each subsection are as follows:

4.a) Collected the titles of all movies displayed on the page and stored them as a Python list.

4.b) Extracted the user ratings for each movie on the page and compiled them into a list.

```python
def movie_info(soup):
  div_cards = soup.find_all("div", class_="card style_1")
  movie_title_list = []
  ratings_list = []
  links_list = []

  for i in range(0, len(div_cards)):
    div = div_cards[i]
    title = div.h2.a
    rate = div.find("div", class_="user_score_chart")["data-percent"]
    link = title["href"]
    movie_title_list.append(title.get_text())
    ratings_list.append(rate)
    links_list.append(link)

  return movie_title_list, ratings_list,links_list

mov,rate,links = movie_info(soup)
print(mov) #List of Movie Titles(4a)
print(rate) #List of User Ratings(4b)
print(links) #List of href links
```

4.c) Retrieved and stored the complete HTML content for every individual movie page by following each movies link and parsing the corresponding webpage.

```python
def html_content(links):

  html_content_list = []
  for link in links:
    html_con = get_BS("https://www.themoviedb.org"+link)
    html_content_list.append(html_con)
  return html_content_list

html_list = html_content(links)
```

4.d) Extracted the genre information for all listed movies and organized it into a list.

```python
def get_movie_genres(html_list):
    genre = []

    for index, html_content in enumerate(html_list, start=1):
        sub_list = []

        # Check if html_content is not None before proceeding
        if html_content:
            genre_span = html_content.find("span", class_="genres")

            # Check if genre_span is not None before proceeding
            if genre_span:
                a_tags = genre_span.find_all("a")
                for a_tag in a_tags:
                    sub_list.append(a_tag.get_text())

        genre.append(sub_list)

    return genre
genre = get_movie_genres(html_list)
print(genre)
```

4.e) Retrieved the cast details for every movie displayed on the page and compiled the extracted values into a Python list.

```python
def get_movie_cast(html_list):
    cast_list = []

    for index, html_content in enumerate(html_list, start=1):
        sub_list = []

        #Check if html_content is not None before proceeding
        if html_content:
            casts = html_content.find_all('li', class_="card")
            for cast in casts:
                #Check if cast.p is not None before accessing its text
                if cast.p:
                    p_text = cast.p.get_text()
                    sub_list.append(p_text)

        cast_list.append(sub_list)

    return cast_list
cast = get_movie_cast(html_list)
print(cast)
```

5) User defined function that returns a pandas data frame with following data are as follows:

* Implemented a user-defined function that consolidates the extracted movie information into a structured pandas DataFrame.
* Included columns for movie titles, corresponding user ratings, identified genres, and extracted cast lists.
* Returned the assembled DataFrame as the final output, providing a clean tabular representation of all scraped movie attributes.

```python
import pandas as pd
import numpy as np

def data_from_page(page_soup):
    mov_page, rate_page,links = movie_info(page_soup)
    html_content_list = html_content(links)
    genre_page = get_movie_genres(html_content_list)
    cast_page = get_movie_cast(html_content_list)

    data = {
        "Title": mov_page,
        "User Rating": rate_page,
        "Genres": genre_page,
        "Cast": cast_page
    }
    movie_dataframe = pd.DataFrame(data)
    return movie_dataframe
```

6) Scraping the data and combining the dataframes.

6.a) Built a function that automates scraping across pages 1 through 5 of the TMDB movie listings, using the previously defined utilities for retrieving page content, extracting individual movie HTML, and assembling DataFrames.

Generated five separate DataFrames, each corresponding to one page of movie listings, ready for export as CSV files.

```python
def data_from_multiple_pages():
    base_url = "https://www.themoviedb.org/movie"
    data_frames = []
    no_pages = 5
    for i in range(1,no_pages+1):
        page_soup = get_BS(base_url+f"?page={i}")
        movie_dataframe = data_from_page(page_soup)
        data_frames.append(movie_dataframe)
    return data_frames
```

6.b) Merged the five DataFrames produced in the previous step into a single combined dataset representing all movies scraped across the first five pages.

```python
data_frames = data_from_multiple_pages()
combined_dataframe = pd.concat(data_frames, ignore_index=True)
combined_dataframe['User Rating'] = combined_dataframe['User Rating'].replace('0', np.nan)
print(combined_dataframe)

#Export the combined DataFrame to a CSV file
combined_dataframe.to_csv('scrappedmoviedata.csv', index=False)
print("Successfully combined the dataframe and csv is exported")
```
